```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("bar_data.csv")
```

```python
df
```

```python
df["LanguagesWorkedWith"] = df["LanguagesWorkedWith"].apply(lambda x :x.split(";"))
```

```python
s= df["LanguagesWorkedWith"]
```

```python
s
```

```python
s[0]
```

```python
a = []
for i in s:
    for j in i :
        a.append(j)
        
```

```python
a
```

```python
len(a)
```

```python
from collections import Counter
```

```python
count = Counter(a)

count
```

```python
trend_lang=count.most_common(15) 
```

```python
my_lang=sorted(trend_lang,key=lambda x : x[1] )
```

```python
my_lang
```

```python
x,y = zip(*my_lang)
```

```python
fig,ax=plt.subplots(figsize=(10,7))

ax.barh(x,y)
```
